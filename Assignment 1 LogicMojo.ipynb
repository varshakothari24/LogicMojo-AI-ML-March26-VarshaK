Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

# 1. Identify and display all transaction values that appear more than once
duplicates = set([x for x in transactions if transactions.count(x) > 1])
print(f"Duplicate transactions: {list(duplicates)}")

# 2. Detect transactions greater than ₹5000 (Fraud Monitoring)
fraud_check = [t for t in transactions if t > 5000]
print(f"High-value transactions (>₹5000): {fraud_check}")

# 3. Calculate the total spending for the day
total_spending = sum(transactions)
print(f"Total spending for the day: ₹{total_spending}")

# 4. Create a new list containing only unique transaction values
# Converting to a set automatically removes duplicates, then we convert back to a list
unique_transactions = list(set(transactions))
print(f"Unique transactions: {unique_transactions}")

# 5. Sort the unique transaction list in descending order
unique_transactions.sort(reverse=True)
print(f"Unique transactions (Sorted Descending): {unique_transactions}")
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
# Initial dataset with invalid entries
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

valid_ages = []
invalid_count = 0

# 1. & 2. Iterate and store only valid ages
for age in ages:
    if age is not None and age >= 0:
        valid_ages.append(age)
    else:
        invalid_count += 1

# 3. Calculate the average age of valid customers
if len(valid_ages) > 0:
    average_age = sum(valid_ages) / len(valid_ages)
else:
    average_age = 0

# 4. Filter customers whose age is greater than 30
above_30 = [age for age in valid_ages if age > 30]

# Displaying the results
print(f"Cleaned Valid Ages: {valid_ages}")
print(f"Average Age: {average_age:.2f}")
print(f"Customers older than 30: {above_30}")
print(f"Number of invalid entries removed: {invalid_count}")
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided:
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:
1. Combines the product names and their prices into a single data structure representing a table.
2. Compares the price of each product across both platforms.
3. Determines which platform offers the lower price.
4. Calculates the difference between the two prices.
5. Displays the product name, cheaper platform, and price difference.

```python
# Initial Data
products = ["Laptop", "Phone", "Tablet", "Monitor", "Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

print(f"{'Product':<12} | {'Cheaper Platform':<18} | {'Price Difference':<15}")
print("-" * 50)

# 1. Combine data and 2. Compare prices
# Using zip to pair elements from all three lists
for product, amz, flp in zip(products, amazon_prices, flipkart_prices):
    
    # 3. Determine which platform offers the lower price
    if amz < flp:
        cheaper_platform = "Amazon"
        # 4. Calculate the difference
        difference = flp - amz
    elif flp < amz:
        cheaper_platform = "Flipkart"
        difference = amz - flp
    else:
        cheaper_platform = "Same Price"
        difference = 0

    # 5. Display the results
    print(f"{product:<12} | {cheaper_platform:<18} | ₹{difference:<15}")
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
# Initial login data
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]

# 1. Separate login hours into even and odd
even_logins = [h for h in login_hours if h % 2 == 0]
odd_logins = [h for h in login_hours if h % 2 != 0]

# 2. Identify login hours that fall into peak activity time (> 5)
peak_logins = [h for h in login_hours if h > 5]

# 3. Sort the login hours in ascending order
# Using sorted() creates a new list without changing the original order of 'login_hours'
sorted_logins = sorted(login_hours)

# 4. Calculate how many logins occurred during peak hours
peak_count = len(peak_logins)

# Displaying the results
print(f"Even-hour logins: {even_logins}")
print(f"Odd-hour logins: {odd_logins}")
print(f"Peak activity hours (>5): {peak_logins}")
print(f"Sorted login hours: {sorted_logins}")
print(f"Total number of peak hour logins: {peak_count}")
```

Question 5:- An e-commerce company exports its daily order log in the following format.
Each order contains multiple pieces of information:

orders = [
["ORD001","Laptop",75000,"Completed"],
["ORD002","Mouse",500,"Completed"],
["ORD003","Laptop",75000,"Returned"],
["ORD004","Keyboard",1500,"Completed"],
["ORD005","Laptop",75000,"Completed"],
["ORD006","Monitor",20000,"Fraud"],
["ORD007","Mouse",500,"Completed"],
["ORD008","Laptop",75000,"Completed"]
]

Each row represents an order with the following structure:

[Order_ID , Product_Name , Price , Status]

Where:
Order_ID → unique order identifier
Product_Name → product purchased
Price → price of the product
Status → order status (Completed, Returned, Fraud)

Your Task:

Write a Python program to simulate a simple order analysis system using loops, nested loops, conditional statements, and list operations.

Your program must perform the following steps.

1. Display All Orders
Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly.

2. Detect Suspicious Activity
The system suspects fraud when:
the same product appears more than 3 times, OR
an order has status "Fraud"

Write a program that scans the dataset and identifies suspicious products or orders.
You must use nested loops to compare product occurrences.

3. Calculate Total Revenue
Only Completed orders contribute to revenue.
Write code that:
iterates through the dataset
adds the price of all completed orders
prints the total revenue generated for the day.

4. Count Order Types
Using loops and conditional statements, count how many orders fall into each category:

Completed
Returned
Fraud
Store these counts in separate variables.

5. Generate a List of Unique Products
Create a new list containing unique product names present in the dataset.
You must implement this without using the set() function.

6. Product Sales Analysis
For each unique product, determine:
how many times it was ordered
total revenue generated by that product
This requires nested loops.

7. Identify the Most Sold Product
Based on the frequency of orders, determine which product appears the most in the dataset.

8. Remove Fraud Orders
Create a cleaned dataset by removing orders whose status is "Fraud".
Use list methods to store the cleaned data in a new list.

```python
orders = [
    ["ORD001", "Laptop", 75000, "Completed"],
    ["ORD002", "Mouse", 500, "Completed"],
    ["ORD003", "Laptop", 75000, "Returned"],
    ["ORD004", "Keyboard", 1500, "Completed"],
    ["ORD005", "Laptop", 75000, "Completed"],
    ["ORD006", "Monitor", 20000, "Fraud"],
    ["ORD007", "Mouse", 500, "Completed"],
    ["ORD008", "Laptop", 75000, "Completed"]
]

# 1. Display All Orders
print("--- Order Table ---")
for row in orders:
    for item in row:
        print(f"{item:<12}", end=" ")
    print() # New line after each row

# 2. Detect Suspicious Activity
print("\n--- Suspicious Activity Report ---")
suspicious_products = []
for i in range(len(orders)):
    product_name = orders[i][1]
    status = orders[i][3]
    
    # Check for "Fraud" status
    if status == "Fraud":
        print(f"Alert: Order {orders[i][0]} flagged as Fraud.")
    
    # Count occurrences for the > 3 rule using a nested loop
    count = 0
    for j in range(len(orders)):
        if orders[j][1] == product_name:
            count += 1
    
    if count > 3 and product_name not in suspicious_products:
        suspicious_products.append(product_name)
        print(f"Alert: Product '{product_name}' appears more than 3 times.")

# 3. Calculate Total Revenue & 4. Count Order Types
total_revenue = 0
completed_count = 0
returned_count = 0
fraud_count = 0

for order in orders:
    status = order[3]
    price = order[2]
    
    if status == "Completed":
        total_revenue += price
        completed_count += 1
    elif status == "Returned":
        returned_count += 1
    elif status == "Fraud":
        fraud_count += 1

print(f"\nTotal Revenue: ₹{total_revenue}")
print(f"Counts -> Completed: {completed_count}, Returned: {returned_count}, Fraud: {fraud_count}")

# 5. Generate a List of Unique Products (Without set())
unique_products = []
for order in orders:
    product = order[1]
    if product not in unique_products:
        unique_products.append(product)
print(f"\nUnique Products: {unique_products}")

# 6. Product Sales Analysis & 7. Identify Most Sold
print("\n--- Product Sales Analysis ---")
most_sold_product = ""
max_orders = 0

for up in unique_products:
    p_count = 0
    p_revenue = 0
    for order in orders:
        if order[1] == up:
            p_count += 1
            if order[3] == "Completed":
                p_revenue += order[2]
    
    print(f"Product: {up:<10} | Orders: {p_count} | Revenue: ₹{p_revenue}")
    
    # Tracking most sold
    if p_count > max_orders:
        max_orders = p_count
        most_sold_product = up

print(f"\nMost Sold Product: {most_sold_product} ({max_orders} orders)")

```

```python
# 8. Remove Fraud Orders
cleaned_orders = []
for order in orders:
    if order[3] != "Fraud":
        cleaned_orders.append(order)

print(f"\nCleaned Dataset (Fraud Removed):")
for co in cleaned_orders:
    print(co)
```
