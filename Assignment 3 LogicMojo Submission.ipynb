Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data   # Feature matrix (NumPy array)
feature_names = iris.feature_names

# 1. Check shape of feature matrix
print("Feature Matrix Shape:", X.shape)  # (150, 4)

# 2. Compute statistics for each feature
mean = np.mean(X, axis=0)
median = np.median(X, axis=0)
std_dev = np.std(X, axis=0)
variance = np.var(X, axis=0)

# Print results
print("\nFeature Statistics:\n")
for i, feature in enumerate(feature_names):
    print(f"{feature}:")
    print(f"  Mean      = {mean[i]:.3f}")
    print(f"  Median    = {median[i]:.3f}")
    print(f"  Std Dev   = {std_dev[i]:.3f}")
    print(f"  Variance  = {variance[i]:.3f}\n")

# 3. Identify feature with highest variability
max_var_index = np.argmax(variance)
print("Feature with highest variability:", feature_names[max_var_index])
print("Variance:", variance[max_var_index])

# 4. Convert one feature into shape (n,1)
sepal_length = X[:, 0]              # shape (150,)
sepal_length_reshaped = sepal_length.reshape(-1, 1)

print("\nOriginal shape:", sepal_length.shape)
print("Reshaped to (n,1):", sepal_length_reshaped.shape)
```

Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

1. Apply standardization:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2. 

Verify:

```python
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data   # shape (150, 4)

# 1. Compute mean and standard deviation (column-wise)
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

# 2. Apply standardization
X_scaled = (X - mean) / std

# 3. Verify results
scaled_mean = np.mean(X_scaled, axis=0)
scaled_std = np.std(X_scaled, axis=0)

print("Mean after scaling:", scaled_mean)
print("Std after scaling:", scaled_std)
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('titanic')

# 1. Fill missing 'age' with median
df['age'] = df['age'].fillna(df['age'].median())

# 2. Fill missing 'embarked' with mode
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# 3. Drop 'deck' column
df = df.drop(columns=['deck'])

# Verify missing values are handled
print(df.isnull().sum())
```

Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create:
family_size = sibsp + parch + 1
2. Create:
is_alone (binary feature)
3. Compute survival rate by is_alone
4. Explain:
Why engineered features improve ML models

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('titanic')

# 1. Create family_size
df['family_size'] = df['sibsp'] + df['parch'] + 1

# 2. Create is_alone (binary feature)
df['is_alone'] = (df['family_size'] == 1).astype(int)

# 3. Compute survival rate by is_alone
survival_by_alone = df.groupby('is_alone')['survived'].mean()

print("Survival Rate by is_alone:\n")
print(survival_by_alone)
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('tips')

# 1. Compute aggregations

# Average tip by day
avg_tip_by_day = df.groupby('day')['tip'].mean()

# Total bill by gender
total_bill_by_gender = df.groupby('sex')['total_bill'].sum()

print("Average Tip by Day:\n", avg_tip_by_day)
print("\nTotal Bill by Gender:\n", total_bill_by_gender)

# 2. Create new feature: tip_percentage
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100

# 3. Identify day with highest tipping behavior
tip_pct_by_day = df.groupby('day')['tip_percentage'].mean()
best_day = tip_pct_by_day.idxmax()

print("\nAverage Tip % by Day:\n", tip_pct_by_day)
print("\nDay with Highest Tipping Behavior:", best_day)
```

Q9. Advanced Filtering + Subsetting

Context: Training data is often filtered based on business rules.

Dataset: Tips

Tasks:

1. Filter:
total_bill > 20 AND tip < 3
2. Select only relevant columns
3. Analyze:
Is this segment under-tipping?
4. Explain:
How such filtering helps anomaly detection

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('tips')

# 1. Filter: total_bill > 20 AND tip < 3
filtered_df = df[(df['total_bill'] > 20) & (df['tip'] < 3)]

# 2. Select only relevant columns
subset = filtered_df[['total_bill', 'tip', 'tip_percentage']] if 'tip_percentage' in df.columns else filtered_df[['total_bill', 'tip']]

# (Optional) create tip_percentage if not already created
subset['tip_percentage'] = (subset['tip'] / subset['total_bill']) * 100

# 3. Analyze
avg_tip_pct = subset['tip_percentage'].mean()
overall_tip_pct = (df['tip'] / df['total_bill'] * 100).mean()

print("Filtered Segment Size:", len(subset))
print("Avg Tip % (Filtered):", round(avg_tip_pct, 2))
print("Avg Tip % (Overall):", round(overall_tip_pct, 2))
```

Q10. Data Transformation for Model Input (NumPy)

Context: ML models require structured numeric arrays.

Tasks:

1. Create two arrays of shape (3,2)
2. Perform:
Vertical stacking
Horizontal stacking
3. Reshape into (2,6)
4. Explain:
Why reshaping is required in ML pipelines

```python
import numpy as np

# 1. Create two arrays of shape (3,2)
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

print("Array A:\n", A)
print("\nArray B:\n", B)

# 2. Vertical stacking (row-wise)
v_stack = np.vstack((A, B))
print("\nVertical Stack Shape:", v_stack.shape)
print(v_stack)

# 3. Horizontal stacking (column-wise)
h_stack = np.hstack((A, B))
print("\nHorizontal Stack Shape:", h_stack.shape)
print(h_stack)

# 4. Reshape into (2,6)
reshaped = h_stack.reshape(2, 6)
print("\nReshaped to (2,6):")
print(reshaped)
print("New Shape:", reshaped.shape)
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# 1. Load dataset
df = sns.load_dataset('titanic')

# Fill numerical
df['age'] = df['age'].fillna(df['age'].median())

# Fill categorical
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Drop high-missing / irrelevant columns
df = df.drop(columns=['deck'])

# One-hot encoding
df_encoded = pd.get_dummies(df, columns=['sex', 'embarked'], drop_first=True)

numerical_cols = ['age', 'fare', 'sibsp', 'parch']

scaler = StandardScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

# Drop target variable if training separately
X = df_encoded.drop(columns=['survived']).values
y = df_encoded['survived'].values

print("Final Feature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Convert to DataFrame
df = pd.DataFrame(X, columns=feature_names)

# 1. Compute correlation matrix
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix)

# 2. Identify highly correlated features (threshold = 0.9)
threshold = 0.9
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            high_corr_pairs.append((col1, col2))

print("\nHighly Correlated Feature Pairs:\n", high_corr_pairs)

# 3. Drop redundant features (example: drop second feature in each pair)
to_drop = [pair[0] for pair in high_corr_pairs]
df_reduced = df.drop(columns=to_drop)

print("\nRemaining Features After Dropping:\n", df_reduced.columns.tolist())
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import pandas as pd
import seaborn as sns

# Load dataset
df = sns.load_dataset('titanic')

# Fill missing age (needed before transformation)
df['age'] = df['age'].fillna(df['age'].median())

# 1 & 2. Create age_group using .apply()
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(categorize_age)

# 3. Compute survival rate per group
survival_by_age_group = df.groupby('age_group')['survived'].mean()

print("Survival Rate by Age Group:\n")
print(survival_by_age_group)
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# 1. Generate 1000 random error values (mean=0, std=1)
np.random.seed(42)  # for reproducibility
errors = np.random.randn(1000)

# 2. Compute mean and standard deviation
mean_error = np.mean(errors)
std_error = np.std(errors)

print("Mean Error:", round(mean_error, 4))
print("Standard Deviation:", round(std_error, 4))

# 3. Identify outliers using mean ± 2*std
lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error

outliers = errors[(errors < lower_bound) | (errors > upper_bound)]

print("\nLower Bound:", round(lower_bound, 4))
print("Upper Bound:", round(upper_bound, 4))
print("Number of Outliers:", len(outliers))
```

Q15. Business + ML Insight Problem (Tips Dataset)

Context: Convert data analysis into decision-making

Tasks:

1. Create:
tip_percentage
2. Group by:
day and time
3. Identify:
Highest revenue segment
4. Provide:
Recommendation for business strategy
How ML model can use this insight
